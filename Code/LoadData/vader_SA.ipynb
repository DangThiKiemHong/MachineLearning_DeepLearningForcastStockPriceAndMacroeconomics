```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from googletrans import Translator

# Tạo đối tượng phân tích cảm xúc và dịch
analyzer = SentimentIntensityAnalyzer()
translator = Translator()

def translate_text(text, src='vi', dest='en'):
    try:
        # Dịch văn bản từ tiếng Việt sang tiếng Anh
        translation = translator.translate(text, src=src, dest=dest)
        return translation.text
    except Exception as e:
        print(f"Translation failed: {e}")
        return text

def analyze_sentiment_vietnamese(text):
    # Dịch văn bản sang tiếng Anh
    translated_text = translate_text(text)

    # Phân tích cảm xúc của văn bản đã dịch
    sentiment = analyzer.polarity_scores(translated_text)

    return sentiment

# Ví dụ văn bản tiếng Việt
text_vietnamese = "Chứng khoán vẫn là kênh đầu tư hấp dẫn?"

# Phân tích cảm xúc
sentiment_result = analyze_sentiment_vietnamese(text_vietnamese)

print(f"Original Text: {text_vietnamese}")
print(f"Translated Text: {translate_text(text_vietnamese)}")
print(f"Sentiment: {sentiment_result}")
```

```python
import pandas as pd

# Đọc dữ liệu từ file CSV
file_path = "/content/kinhtechungkhoan.csv"
df = pd.read_csv(file_path, encoding='utf8')  # Hoặc encoding='cp1252'

# Dịch và phân tích cảm xúc cho từng dòng văn bản
df['translated_text'] = df['title'].apply(translate_text)
df['sentiment'] = df['translated_text'].apply(analyze_sentiment_vietnamese)
sentiment_df = pd.DataFrame(df['sentiment'].tolist())
sentiment_df.columns = ['neg_new', 'neu_new', 'pos_new', 'compound_new']

df = df.join(sentiment_df)


# Hiển thị kết quả
print(df.head())
# Lưu kết quả vào file CSV mới
output_file_path = "/content/kinhtechungkhoan_output.csv"
df.to_csv(output_file_path, index=False)

print(f"Results saved to {output_file_path}")
```
